```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('results/results_whitebox_20-71.csv')
#df = pd.read_csv('results/results_whitebox_20-91.csv')
#df = pd.read_csv('results/results_whitebox_50-218.csv')
#df = pd.read_csv('results/results_whitebox_75-325_v3.csv')
df['ratio'] = df.best_cost / df.optimum
df['found'] = df.best_cost == df.optimum
df
```

```python
grouped = df.groupby(['temperature', 'alpha']).mean()[['ratio', 'found', 'iter']]

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(font_scale=3)
```

```python
heatmap = grouped.reset_index().pivot(index='temperature', columns='alpha')['found']
fig, ax = plt.subplots(figsize=(12, 12),)
#ax.set_title('no. of runs that found optimum vs all runs ratio')
fig = sns.heatmap(heatmap, annot=True, annot_kws={"fontsize": 30}, cmap="Reds")
```

```python
heatmap = grouped.reset_index().pivot(index='temperature', columns='alpha')['iter']
fig, ax = plt.subplots(figsize=(12, 12), )
fig = sns.heatmap(heatmap, annot=True, annot_kws={"fontsize": 30}, cmap="Reds")
```
